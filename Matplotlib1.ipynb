```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,12,250)
```

```python
y = np.sin(x)
```

```python
plt.figure(figsize = (4,2))
plt.plot (x , y,"--g")

plt.xlabel("this is my x axis")
plt.ylabel("this is y axis")
plt.title("this is title")
plt.grid()
plt.show()
```

```python
n = np.random.rand(50)
ny = np.random.rand(50)
```

```python
plt.figure(figsize = (4,2))
colors = np.random.rand(50)
size = 1000*np.random.rand(50)
plt.scatter(n,ny,c = colors,s = size,alpha= 0.65,marker = ".")
```

#### Plotting in a 3d plane

```python
n = np.random.rand(50)
ny = np.random.rand(50)
nz = np.random.rand(50)
```

```python
fig = plt.figure()

## figure in 3d projection
ax = fig.add_subplot(projection = "3d")
ax.scatter(n,ny,nz)
plt.show()
```
